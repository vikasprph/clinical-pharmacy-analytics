```python
import pandas as pd

df = pd.read_csv("pharmacy_claims_final.csv")
df.head()
```

```python
def recommendation(row):
    if row["drug_name"] == "Crestor":
        return "Switch to generic statin"
    elif row["drug_name"] == "Diovan":
        return "Switch to generic valsartan"
    elif row["drug_name"] == "Humira":
        return "Evaluate biosimilar option"
    elif row["drug_name"] == "Stelara":
        return "Specialty utilization review"
    elif row["drug_name"] == "Lantus":
        return "Evaluate lower-cost insulin alternative"
    elif row["drug_name"] == "Ozempic":
        return "Assess formulary alternatives"
    elif row["drug_name"] in ["Eliquis", "Xarelto"]:
        return "Review anticoagulant utilization"
    else:
        return "No immediate recommendation"

df["recommendation"] = df.apply(recommendation, axis=1)

df[["drug_name", "claim_cost", "estimated_savings", "recommendation"]].head(20)
```

```python
recommendation_summary = (
    df.groupby("recommendation")["estimated_savings"]
    .sum()
    .sort_values(ascending=False)
)

recommendation_summary

```

```python
df.to_csv("pharmacy_claims_with_recommendations.csv", index=False)
```

```python
import pandas as pd

df = pd.read_csv("pharmacy_claims_with_recommendations.csv")
```

```python
def priority_level(row):
    if row["estimated_savings"] > 1000:
        return "High Priority"
    elif row["estimated_savings"] > 200:
        return "Medium Priority"
    else:
        return "Low Priority"

df["priority_level"] = df.apply(priority_level, axis=1)
```

```python
df[["drug_name", "estimated_savings", "priority_level"]].head(20)
```

```python
df.groupby("priority_level")["estimated_savings"].sum()
```

```python
df.to_csv("pharmacy_claims_final_v2.csv", index=False)
```
