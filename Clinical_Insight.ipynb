```python
# =========================
# Load Pharmacy Claims Data
# =========================

import pandas as pd

df = pd.read_csv("pharmacy_claims_final.csv")
df.head()
```

# =========================
# Clinical Opportunity Analysis
# =========================

```python
def recommendation(row):
    if row["drug_name"] == "Crestor":
        return "Switch to generic statin"
    elif row["drug_name"] == "Diovan":
        return "Switch to valsartan"
    elif row["drug_name"] == "Humira":
        return "Evaluate biosimilar"
    elif row["drug_name"] == "Stelara":
        return "Specialty drug review"
    elif row["drug_name"] == "Lantus":
        return "Consider lower-cost insulin"
    elif row["drug_name"] == "Ozempic":
        return "Assess formulary alternatives"
    else:
        return "No change"

df["recommendation"] = df.apply(recommendation, axis=1)
```

```python
def priority(row):
    if row["estimated_savings"] > 1000:
        return "High Priority"
    elif row["estimated_savings"] > 200:
        return "Medium Priority"
    else:
        return "Low Priority"

df["priority_level"] = df.apply(priority, axis=1)
```

```python
df.groupby("recommendation")["estimated_savings"].sum().sort_values(ascending=False)
```

```python
def generate_report(df):
    total_spend = df["claim_cost"].sum()
    total_savings = df["estimated_savings"].sum()

    print("===== Pharmacy Analytics Report =====")
    print(f"Total Spend: ${total_spend:,.0f}")
    print(f"Estimated Savings: ${total_savings:,.0f}")

    print("\nTop Cost Drivers:")
    print(df.groupby("drug_name")["claim_cost"].sum().sort_values(ascending=False).head(5))

    print("\nTop Savings Opportunities:")
    print(df.groupby("recommendation")["estimated_savings"].sum().sort_values(ascending=False).head(5))

generate_report(df)
```
