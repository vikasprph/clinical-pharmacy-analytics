```python
import pandas as pd

df = pd.read_csv("pharmacy_claims_v2.csv")
df.head()
```

```python
def generate_pharmacy_report(df):
    total_spend = df["claim_cost"].sum()
    total_savings = df["estimated_savings"].sum()

    print("CLINICAL PHARMACY ANALYTICS REPORT")
    print("=" * 45)
    print(f"Total Pharmacy Spend: ${total_spend:,.0f}")
    print(f"Estimated Cost Savings: ${total_savings:,.0f}")
    print()

    print("Top 5 Cost Drivers:")
    print(
        df.groupby("drug_name")["claim_cost"]
        .sum()
        .sort_values(ascending=False)
        .head(5)
    )
    print()

    print("Savings by Clinical Recommendation:")
    print(
        df.groupby("recommendation")["estimated_savings"]
        .sum()
        .sort_values(ascending=False)
    )
    print()

    print("Savings by Priority Level:")
    print(
        df.groupby("priority_level")["estimated_savings"]
        .sum()
        .sort_values(ascending=False)
    )

generate_pharmacy_report(df)
```

```python
top_cost_drivers = (
    df.groupby("drug_name")["claim_cost"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

savings_by_recommendation = (
    df.groupby("recommendation")["estimated_savings"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

savings_by_priority = (
    df.groupby("priority_level")["estimated_savings"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

top_cost_drivers.to_csv("top_cost_drivers.csv", index=False)
savings_by_recommendation.to_csv("savings_by_recommendation.csv", index=False)
savings_by_priority.to_csv("savings_by_priority.csv", index=False)
```
